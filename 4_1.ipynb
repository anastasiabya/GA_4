```python
import numpy as np
import pandas as pd
import pygad
```

```python
df = pd.read_csv('12.txt', delimiter=' ', engine='python')
df
```

```python
parameters = df.columns.tolist()
W = int(parameters[0]) #Грузоподъемность
V = int(parameters[1]) #Вместимость
df.columns = ['v', 'c']
df['w'] = df.index
data = df[['w','v','c']].values.tolist()
```

```python
data
```

```python
def fitness_func(solution, solution_idx):
    w = 0 #Вес
    v = 0 #Объем
    c = 0 #Ценность
    for (array, item) in zip(solution, data):
        if array > 0:
            w += item[0]
            v += item[1]
            c += item[2]
    if w > W or v > V:
        c = 0
    return c
```

```python
ga_instance = pygad.GA(num_generations=200, #Количество поколений
                       num_parents_mating=2, #Количество решений, которые будут выбраны в качестве родителей
                       fitness_func=fitness_func, #Функция, вычисляющая значение пригодности решения
                       sol_per_pop=10, #Количество решений в популяции
                       num_genes=len(data), #Количество генов в решении
                       init_range_low=-1, #Нижнее значение случайного диапазона, из которого выбираются значения генов в исходной популяции
                       init_range_high=1 #Верхнее значение случайного диапазона, из которого выбираются значения генов в исходной популяции
                       )
```

```python
ga_instance.run()
solution, solution_fitness, solution_idx = ga_instance.best_solution()
```

```python
result = []
for i in range(0, len(data)):
    if solution[i] > 0:
        result.append(data[i])
result_df = pd.DataFrame.from_records(result, columns=['w','v','c'])
```

```python
print(solution_fitness)
print(result_df)
w = 0
v = 0
c = 0
for i in range(0, len(result_df)):
        w += result_df.at[i,'w']
        v += result_df.at[i,'v']
        c += result_df.at[i,'c']
print("Вес:", w, "(Грузоподъемноcть:", W, ")")
print("Объем:", round(v,1), "(Вместимость:", V, ")")
print("Ценность:", c)
```

```python
file=open('4_1.txt','w', encoding='utf-8')
file.write(result_df.to_string())
file.write("\nВес: " + str(w) + " (Грузоподъемноcть: " + str(W) + ")")
file.write("\nОбъем: " + str(round(v,1)) + " (Вместимость: " + str(V) + ")")
file.write("\nЦенность: " + str(c))
file.close()
```
